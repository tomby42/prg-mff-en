```python
import numpy as np
import matplotlib.pyplot as plt
import math

plt.rcParams['figure.figsize'] = [15, 10]

def draw_fn(f, rng = (-10, 10, 100)):
    x = np.linspace(*rng)
    y = list(map(f, x))
    plt.plot(x,y)
    plt.grid()
    plt.axvline()
    plt.axhline()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

__Sets__

- Collection of elements that are different
- Contruction forbids self reference. See Russell's paradox: R = {x | x not ∈ x} R ∈ R if and only if R not ∈ R
- 0 or {} - empty set
- basic operation
  - union, intersection, complement, products, ... 

N - Set of natural numbers: 0, 1, 2, ....
- construction 0 -> {}, 1 -> {{}, {{}}}, n -> {n,{n}}

Z - Set of integers: ..., -2, -1, 0, 1, 2, ....

R - Set of real numbers

__Binary Relations__

Sets X,Y: R ⊆ X x Y, {(x, y) ∈ R, x ∈ X, y ∈ Y}

function: relation R, x R y & x R z -> y = z => table with unique keys

__Polynomials__

f: R -> R

f(x) = a<sub>n</sub>x<sup>n</sup> + a<sub>n-1</sub>x<sup>n-1</sup>+....+a<sub>0</sub>x<sup>0</sup>

```python
draw_fn(lambda x: 2*x**3 - 3*x**2 - 5*x + 4)
```

__Logarithms__

log(x): R<sup>+</sup> -> R

log<sub>b</sub>(uv) = log<sub>b</sub> u + log<sub>b</sub> v

log<sub>b</sub>(u / v) = log<sub>b</sub> u - log<sub>b</sub> v

log<sub>b</sub>(u<sup>p</sup>) = p log<sub>b</sub> u

```python
draw_fn(lambda x: math.log10(x), rng = (1e-2, 100, 1000))
```
